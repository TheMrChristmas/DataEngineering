```python
from pathlib import Path

import pandas as pd

candidates = [
    Path("/opt/airflow/data/raw/yellow_tripdata_2026-01.parquet"),  # inside Airflow container
    Path("../../data/raw/yellow_tripdata_2026-01.parquet"),          # from dags/src notebook location
    Path("data/raw/yellow_tripdata_2026-01.parquet"),
 ]

parquet_path = next((p for p in candidates if p.exists()), None)
if parquet_path is None:
    tried = [str(p.resolve()) for p in candidates]
    raise FileNotFoundError(f"Could not find parquet file. Tried: {tried}")

df = pd.read_parquet(parquet_path)
print(f"Using file: {parquet_path.resolve()}")
print(df.dtypes)
print("---")
print(df.head(3))
print("---")
print(df.isnull().sum())
```
